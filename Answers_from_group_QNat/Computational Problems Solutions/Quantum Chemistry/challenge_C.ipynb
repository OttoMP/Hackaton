##Step 1:

Crearting a function that implements the evolution operator related to the Hamiltonian
$$H = \alpha X \otimes X + \beta Z \otimes Z.$$

Using,
$$e^{(A+B)} = \text{lim}_{N → ∞} (e^{A/n}e^{B/n})^N.$$
So
$$U(t) \approx (e^{-i α X \otimes X t/N} e^{-i β Z \otimes Z t/N})^N$$

The exponential $e^{-i φ X \otimes X }$ is known as IsingXX gate, and $e^{-i φ Z \otimes Z }$ is known as IsingZZ gate. So we can use the `qml.IsingXX` and `qml.IsingZZ` to implement this gates (As was made in the function `troterization()`) or implement gate by gate as was made in `troterization1()`.

```python
num_qubits = 2

dev = qml.device('default.qubit',wires=num_qubits)

@qml.qnode(dev)
def troterization(alpha, beta, time, troter_steps):
  for i in range(troter_steps):
    qml.IsingXX(2*alpha*time/troter_steps, wires = [0,1])
    qml.IsingZZ(2*beta*time/troter_steps, wires = [0,1])

  return qml.probs(wires = [0,1])
```

#Creating The IsingZZ and IsingXX

```python
num_qubits = 2

dev = qml.device('default.qubit',wires=num_qubits)

@qml.qnode(dev)
def troterization1(alpha, beta, time, troter_steps):
  for i in range(troter_steps):

    #=============== Ising XX gate ===============#
    qml.CNOT(wires= [0,1])
    qml.RX(2*alpha*time/troter_steps, wires = [0])
    qml.CNOT(wires= [0,1])
    #=============== Ising ZZ gate ===============#
    qml.CNOT(wires= [0,1])
    qml.RZ(2*beta*time/troter_steps, wires = [1])
    qml.CNOT(wires= [0,1])

  return qml.probs(wires = [0,1])
```
